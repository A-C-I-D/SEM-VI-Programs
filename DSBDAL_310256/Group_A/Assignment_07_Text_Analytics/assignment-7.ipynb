BLOCK 1: IMPORTS AND SETUP

```python
import nltk
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from collections import Counter

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag


print("NLP PREPROCESSING PIPELINE WITH TF-IDF")
```

BLOCK 2: SAMPLE DOCUMENTS (Kaggle-style dataset simulation)

```python
documents = [
    "The quick brown fox jumps over the lazy dog. Running and jumping are fun activities!",
    "Machine learning is fascinating. Learning algorithms improve with more data and training.",
    "Natural language processing enables computers to understand human language effectively.",
    "The fox was running quickly through the forest. It jumped over the fallen tree.",
    "Deep learning models require large amounts of training data for better performance."
]

print("📄 SAMPLE DOCUMENTS:")
print("-" * 60)
for i, doc in enumerate(documents, 1):
    print(f"Document {i}: {doc[:80]}...")
print(f"\nTotal Documents: {len(documents)}")
```

BLOCK 3: TOKENIZATION

```python
print("\n" + "=" * 60)
print("STEP 1: TOKENIZATION")
print("=" * 60)

tokenized_docs = []
for i, doc in enumerate(documents, 1):
    tokens = word_tokenize(doc.lower())
    tokens = [token for token in tokens if token.isalpha()]
    tokenized_docs.append(tokens)
    print(f"\nDoc {i} Tokens ({len(tokens)} tokens):")
    print(tokens[:15], "..." if len(tokens) > 15 else "")
```

BLOCK 4: POS TAGGING (Part-of-Speech)

```python
print("STEP 2: POS TAGGING")

pos_tagged_docs = []
for i, tokens in enumerate(tokenized_docs, 1):
    tagged = pos_tag(tokens)
    pos_tagged_docs.append(tagged)
    print(f"\nDoc {i} POS Tags (first 10):")
    print(tagged[:10])
```

BLOCK 5: STOP WORDS REMOVAL

```python
print("STEP 3: STOP WORDS REMOVAL")

stop_words = set(stopwords.words('english'))
filtered_docs = []

for i, tokens in enumerate(tokenized_docs, 1):
    filtered = [token for token in tokens if token not in stop_words]
    filtered_docs.append(filtered)
    removed = len(tokens) - len(filtered)
    print(f"\nDoc {i}:")
    print(f"  Before: {len(tokens)} tokens")
    print(f"  After:  {len(filtered)} tokens")
    print(f"  Removed: {removed} stop words")
    print(f"  Sample: {filtered[:12]}")
```

BLOCK 6: STEMMING (Porter Stemmer)

```python
print("STEP 4: STEMMING")

stemmer = PorterStemmer()
stemmed_docs = []

for i, tokens in enumerate(filtered_docs, 1):
    stemmed = [stemmer.stem(token) for token in tokens]
    stemmed_docs.append(stemmed)
    print(f"\nDoc {i} Stemmed (first 12):")
    print(stemmed[:12])
    
    # Show stemming examples
    examples = [(t, s) for t, s in zip(tokens[:6], stemmed[:6]) if t != s]
    if examples:
        print(f"  Stemming examples: {examples}")
```

BLOCK 7: LEMMATIZATION

```python
print("STEP 5: LEMMATIZATION")

lemmatizer = WordNetLemmatizer()
lemmatized_docs = []

for i, tokens in enumerate(filtered_docs, 1):
    lemmatized = [lemmatizer.lemmatize(token) for token in tokens]
    lemmatized_docs.append(lemmatized)
    print(f"\nDoc {i} Lemmatized (first 12):")
    print(lemmatized[:12])
    
    examples = [(t, l) for t, l in zip(tokens[:6], lemmatized[:6]) if t != l]
    if examples:
        print(f"  Lemmatization examples: {examples}")
```

BLOCK 8: COMPARE STEMMING vs LEMMATIZATION

```python
print("COMPARISON: STEMMING vs LEMMATIZATION")

comparison_df = pd.DataFrame({
    'Original': filtered_docs[0][:10],
    'Stemmed': stemmed_docs[0][:10],
    'Lemmatized': lemmatized_docs[0][:10]
})
print(comparison_df.to_string(index=False))
```

BLOCK 9: TF-IDF VECTORIZATION (Using Lemmatized Text)

```python
print("STEP 6: TF-IDF REPRESENTATION")

processed_documents = [' '.join(tokens) for tokens in lemmatized_docs]

print("\nProcessed Documents for TF-IDF:")
for i, doc in enumerate(processed_documents, 1):
    print(f"Doc {i}: {doc[:100]}...")

# Initialize TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(
    max_features=50,      
    min_df=1,             
    max_df=1.0,           
    ngram_range=(1, 1)   
)

tfidf_matrix = tfidf_vectorizer.fit_transform(processed_documents)

print(f"\n TF-IDF Matrix Shape: {tfidf_matrix.shape}")
print(f"   Documents: {tfidf_matrix.shape[0]}")
print(f"   Features (Terms): {tfidf_matrix.shape[1]}")
```

BLOCK 10: TF-IDF FEATURE NAMES AND MATRIX

```python
print("TF-IDF FEATURE NAMES (Vocabulary)")

feature_names = tfidf_vectorizer.get_feature_names_out()
print(f"Total Features: {len(feature_names)}")
print(f"\nFeature Names:\n{feature_names}")

tfidf_dense = tfidf_matrix.todense()
```

BLOCK 11: TF-IDF DATAFRAME VISUALIZATION

```python
print("TF-IDF MATRIX (Document-Term Matrix)")

tfidf_df = pd.DataFrame(
    tfidf_dense,
    columns=feature_names,
    index=[f'Doc_{i}' for i in range(1, len(documents) + 1)]
)

# Display full matrix
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
print(tfidf_df.round(4))
```

BLOCK 12: TOP TF-IDF SCORES PER DOCUMENT

```python
print("TOP 5 TF-IDF SCORES PER DOCUMENT")

for i in range(len(documents)):
    doc_name = f'Doc_{i+1}'
    row = tfidf_df.iloc[i]
    # Get non-zero scores
    non_zero = row[row > 0].sort_values(ascending=False).head(5)
    print(f"\n{doc_name} - Top Terms:")
    for term, score in non_zero.items():
        print(f"  {term:15} : {score:.4f}")
```

BLOCK 13: MANUAL TF-IDF CALCULATION 

```python
print("MANUAL TF-IDF CALCULATION (Understanding the Math)")

def calculate_tf(document):
    """Calculate Term Frequency"""
    word_count = Counter(document)
    total_words = len(document)
    tf = {word: count/total_words for word, count in word_count.items()}
    return tf

def calculate_idf(documents):
    """Calculate Inverse Document Frequency"""
    N = len(documents)
    idf = {}
    all_words = set(word for doc in documents for word in doc)
    
    for word in all_words:
        doc_count = sum(1 for doc in documents if word in doc)
        idf[word] = np.log(N / doc_count) + 1  # Smoothed IDF
    
    return idf

print("\n Manual Calculation for Document 1:")
tf_doc1 = calculate_tf(lemmatized_docs[0])
print(f"\nTerm Frequencies (TF) - Top 5:")
for word, score in sorted(tf_doc1.items(), key=lambda x: x[1], reverse=True)[:5]:
    print(f"  {word:12} : {score:.4f}")

idf = calculate_idf(lemmatized_docs)
print(f"\nInverse Document Frequencies (IDF) - Sample:")
sample_words = list(idf.keys())[:5]
for word in sample_words:
    print(f"  {word:12} : {idf[word]:.4f}")

print(f"\nManual TF-IDF for 'fox' in Doc 1:")
tf_fox = tf_doc1.get('fox', 0)
idf_fox = idf.get('fox', 0)
tfidf_fox = tf_fox * idf_fox
print(f"  TF('fox')    = {tf_fox:.4f}")
print(f"  IDF('fox')   = {idf_fox:.4f}")
print(f"  TF-IDF('fox')= {tfidf_fox:.4f}")
```

BLOCK 14: FINAL SUMMARY

```python
print("PIPELINE SUMMARY")

summary_data = {
    'Step': ['Original', 'Tokenization', 'POS Tagging', 'Stop Words Removal', 
             'Stemming', 'Lemmatization', 'TF-IDF Vectorization'],
    'Description': [
        'Raw text documents',
        'Split into tokens, lowercase, remove punctuation',
        'Assign grammatical categories (NN, VB, etc.)',
        'Remove common words (the, is, at, etc.)',
        'Reduce words to root form (running → run)',
        'Reduce to dictionary form (better → good)',
        'Numerical representation of text importance'
    ],
    'Output': [
        f'{len(documents)} documents',
        f'{sum(len(d) for d in tokenized_docs)} total tokens',
        f'{sum(len(d) for d in pos_tagged_docs)} tagged tokens',
        f'{sum(len(d) for d in filtered_docs)} filtered tokens',
        f'{sum(len(d) for d in stemmed_docs)} stemmed tokens',
        f'{sum(len(d) for d in lemmatized_docs)} lemmatized tokens',
        f'{tfidf_matrix.shape} matrix'
    ]
}

summary_df = pd.DataFrame(summary_data)
print(summary_df.to_string(index=False))

print(" NLP PREPROCESSING PIPELINE COMPLETED!")
```
