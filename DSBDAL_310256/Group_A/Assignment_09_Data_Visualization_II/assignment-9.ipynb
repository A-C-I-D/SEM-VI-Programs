```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')
```

```python
# Create a box plot for age distribution by gender and survival status
plt.figure(figsize=(12, 8))

# Create the box plot with hue for survival status
sns.boxplot(data=titanic, x='sex', y='age', hue='survived', palette='Set2')

plt.title('Age Distribution by Gender and Survival Status', fontsize=16, fontweight='bold')
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Age (years)', fontsize=12)

plt.legend(title='Survival Status', labels=['Did Not Survive', 'Survived'])

plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))

# Create violin plot with split for survival status
sns.violinplot(data=titanic, x='sex', y='age', hue='survived', 
               split=True, palette='Set2', inner='quartile')

plt.title('Age Distribution Density by Gender and Survival Status', fontsize=16, fontweight='bold')
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Age (years)', fontsize=12)
plt.legend(title='Survival Status', labels=['Did Not Survive', 'Survived'])

plt.tight_layout()
plt.show()
```

```python
# Statistical summary
print("Age Statistics by Gender and Survival:")
print("="*50)
summary = titanic.groupby(['sex', 'survived'])['age'].describe()
print(summary)
```
